```python
from phylodeep import BD, BDEI, BDSS, FULL
from phylodeep.checkdeep import checkdeep
from phylodeep.modeldeep import modeldeep
from phylodeep.paramdeep import paramdeep

path_to_tree = "Zurich_test/HIV_tree.trees"

# set presumed sampling probability
sampling_proba = 0.25


#a priori check for models 
checkdeep(path_to_tree, model=BD, outputfile_png="Zurich_test/BD_check.png")
checkdeep(path_to_tree, model=BDEI, outputfile_png="Zurich_test/BDEI_check.png")
checkdeep(path_to_tree, model=BDSS, outputfile_png="Zurich_test/BDSS_check.png")

```

```python
# model selection
model_BDEI_vs_BD_vs_BDSS = modeldeep(path_to_tree, sampling_proba, vector_representation= FULL)
```

```python
model_BDEI_vs_BD_vs_BDSS
```

```python
# parameter inference 
param_BDSS = paramdeep(path_to_tree, sampling_proba, model=BDSS, vector_representation=FULL, ci_computation=True)
param_BDSS
```
